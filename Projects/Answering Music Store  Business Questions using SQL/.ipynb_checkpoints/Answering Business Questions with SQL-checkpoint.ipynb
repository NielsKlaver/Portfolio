# Music store Business questions

In this project I explore the database of a music store to recommend which albums or songs to add.

```python
import sqlite3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

db = 'chinook.db'
```

Create functions for reusability

```python
def run_query(q):
    with sqlite3.connect(db) as conn:
        return pd.read_sql(q, conn)

def run_command(c):
    with sqlite3.connect(db) as conn:
        conn.isolation_level = None
        conn.execute(c)

def show_tables():
    q = '''SELECT
           name,
           type
           FROM sqlite_master
           WHERE type IN ("table","view"); '''
    return run_query(q)

show_tables()
```

__Wat is the most populair genre in the USA ?__

```python
genre_usa = ''' 
        WITH
        tracks_sold_usa AS (SELECT il.track_id, c.country FROM invoice_line il
        INNER JOIN invoice i ON i.invoice_id = il.invoice_id
        INNER JOIN customer c ON c.customer_id = i.customer_id
        WHERE c.country = "USA"
        ),
        track_genre AS 
        (SELECT t.track_id , g.name FROM genre g
        INNER JOIN track t ON t.genre_id = g.genre_id)
        
        SELECT tsu.country Country, 
        tg.name Genre ,
        COUNT (tg.name) Purchaches ,
        CAST( Count(tg.name) AS Float) / 
        (SELECT Count (*) from tracks_sold_usa) Percentage
        
        FROM tracks_sold_usa tsu
        INNER JOIN track_genre tg on tg.track_id = tsu.track_id 
        GROUP BY tg.name
        ORDER BY 3  DESC
        '''
    
run_query(genre_usa) 
```

```python
run_query(genre_usa).plot(kind='bar', x='Genre', y= 'Purchaches')
```

The most populair genre in the usa is Rock followed by Alternative & Punk, Metal and R&B /Soul




__Compare sales employees__

```python
sales_per_emp = '''WITH sales_per_emp AS ( 
                        SELECT c.customer_id ,c.support_rep_id, SUM(i.total) sales,
                        e.last_name name
                        FROM customer c
                        INNER JOIN invoice i ON i.customer_id = c.customer_id
                        INNER JOIN employee e on e.employee_id = c.support_rep_id
                        GROUP BY 1) 
                        
                        SELECT spe.name, sum(spe.sales) total_sales
                        FROM sales_per_emp spe
                        GROUP BY 1; '''
run_query(sales_per_emp)
```

```python
run_query(sales_per_emp).plot(kind='bar', y= 'total_sales', x = 'name')
```

## Analyse sales per country

I need to combine all countries with only 1 customer in "Other" category. Thereafter calculate average sales per customer and total sales.
I calculate the value of each costumer sorted by country

```python
spc = ''' WITH country_other AS (
          SELECT CASE
                 WHEN (SELECT count(*)
                 FROM customer
                 WHERE country = c.country) = 1 
                 THEN "Other"
                 ELSE c.country
                 END AS country,
                 c.customer_id, il.*
                 FROM invoice_line il
         INNER JOIN invoice i ON i.invoice_id = il.invoice_id
         INNER JOIN customer c ON c.customer_id = i.customer_id )

SELECT
    country,
    customers,
    average_order,
    total_sales,
    customer_value
FROM
    (SELECT country,
        COUNT(DISTINCT customer_id) customers,
        SUM(unit_price) total_sales,
        ROUND(SUM(unit_price) / count(distinct customer_id),2)
        customer_value,
        ROUND(SUM(unit_price) / count(distinct invoice_id),2)
        average_order,
        CASE
            WHEN country = "Other" THEN 1
            ELSE 0
        END AS other
    FROM country_other
    GROUP BY country
    ORDER BY other ASC, total_sales DESC
    ); '''


run_query(spc)
```

Since this is an English webshop it looks like there is potential in United Kingdom and India to increase sales. Also the customer value in the Czech republic needs further investigation. This could be a potentially lucrative market or a anomaly because of the small number of customers.

__Analyze if customers bought entire albums or just songs__

Compare all invoice_id s to look if they bought an album or not. I check for every invoice one song to get the correct album. Afterwards I will check if all songs of this album are on the invoice an use the except function to check if it returns NULL . Then count album purchases as percentage of total invoices.

```python
q_tracks = '''WITH invoice_first_track AS (
                 SELECT  il.invoice_id invoice_id,
                 MIN(il.track_id) first_track_id
                 FROM invoice_line il
                 GROUP BY 1    )

        SELECT
        album_purchase,
        COUNT(invoice_id) number_of_invoices,
        CAST(count(invoice_id) AS FLOAT) /
        ( SELECT COUNT(*) FROM invoice) percent
        FROM   ( SELECT
        ifs.*,
        CASE
            WHEN
                 (SELECT t.track_id FROM track t
                  WHERE t.album_id = ( SELECT t2.album_id FROM track t2
                                      WHERE t2.track_id = ifs.first_track_id ) 

                  EXCEPT 

                  SELECT il2.track_id FROM invoice_line il2
                  WHERE il2.invoice_id = ifs.invoice_id
                 ) IS NULL
             AND
                 (SELECT il2.track_id FROM invoice_line il2
                  WHERE il2.invoice_id = ifs.invoice_id

                  EXCEPT 

                  SELECT t.track_id FROM track t
                  WHERE t.album_id = ( SELECT t2.album_id FROM track t2
                                     WHERE t2.track_id = ifs.first_track_id ) 
                                      ) IS NULL
             THEN "album"
             ELSE "songs"
         END AS "album_purchase"
     FROM invoice_first_track ifs
    )
GROUP BY album_purchase;'''

run_query(q_tracks)
```

Most of the purchases where songs and not albums. The shop might want to consider buying just the most populair songs of every album
